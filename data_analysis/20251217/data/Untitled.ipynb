```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
Rf = 1000
Cf = 100*1e-12
Ci = 23*1e-12
Aol = 10000
Iin = 50e-6
```

```python
freq = np.logspace(0, 5, 100)
s = 2j * np.pi * freq
beta = (1 + s * Rf * Cf) / (1 + s * Rf * (Ci + Cf))
```

```python
Vout = -Iin * Rf / (1 + 1 / (Aol * beta))
```
