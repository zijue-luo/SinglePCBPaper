```python
def trimmed_stats(x, trim = 0.10):
    trim_len = int(len(x) * trim)
    trimmed_x = np.sort(x)[trim_len:-trim_len]
    return np.mean(trimmed_x), np.std(trimmed_x)
```

```python
def avg_below_pct(y, pct):
    """
    Guess baseline value using the average of part of the lowest data.
    """

    y = np.asarray(y)
    y = y[np.isfinite(y)]
    n = y.size
    if n == 0:
        return np.nan

    pct = float(pct)
    pct = np.clip(pct, 0.0, 1.0)

    k = max(1, int(np.ceil(n * pct)))
    low = np.partition(y, k-1)[:k]
    
    return low.mean()
```

```python
def baseline_edges(y, edge_frac=0.10, edge_min=5):
    """
    Guess baseline value using the median of edges.
    """
    
    n = len(y)
    k = max(edge_min, int(n * edge_frac))
    return float(np.median(np.r_[y[:k], y[-k:]]))
```

# Data Loading, Preview and Dumping

```python
def load_data(timestamp):

    basepath = os.path.abspath(r"C:\Users\lluoz\Documents\ZJLalala\Research\paper_writing\data_analysis\20251217\data")
    date, time = timestamp.split("_")

    # Read Main Data
    data_to_read = [
        "act_RF_amplitude", "last_frequency_422", "last_frequency_390",  # status raw data
        "arr_of_setpoints",                                              # x data
        "trapped_signal", "lost_signal", "ratio_signal", "ratio_lost"    # y data
    ]
    data = {}
    for entry in data_to_read:
        filename = f"{timestamp}_{entry}"
        filepath = os.path.join(basepath, date, filename)
        data[entry] = np.genfromtxt(filepath, delimiter=",")

    conf_path = os.path.join(basepath, date, f"{timestamp}_conf")
    with open(conf_path) as conf_file:
        config_text = conf_file.readlines()

    # Read Experiment Configs
    data["config"] = {}
    for i, line in enumerate(config_text):
        if line[0] == "[":
            name = line[1:-2]
            next_line = config_text[i+1]
            if next_line[:6] == "val = ":
                value = next_line[6:-1]
                data["config"][name] = value
    
    #print(data["config"])

    # Read Measured RF amplitude and Laser Frequencies
    status_to_read = ["act_RF_amplitude", "last_frequency_422", "last_frequency_390"]
    data["status"] = {}
    for entry in status_to_read:
        filename = f"{timestamp}_{entry}"
        filepath = os.path.join(basepath, date, filename)
        raw_data = np.genfromtxt(filepath, delimiter=",")
        mean, std = trimmed_stats(raw_data)
        data["status"][entry] = {"mean": mean, "std": std}

    print(data["status"])

    return data
```

```python
def plot_data(data_dict, ratio = True):

    fig, (ax1, ax2) = plt.subplots(2, 1, figsize=(10, 10))

    x = data_dict["arr_of_setpoints"]
    if ratio:
        y1 = data_dict["ratio_signal"]
        y2 = data_dict["ratio_lost"]
        ylabel1 = "Trapped Count / Loading Count"
        ylabel2 = "Lost Count / Loading Count"
    else:
        y1 = data_dict["trapped_signal"]
        y2 = data_dict["lost_lost"]
        ylabel1 = "Trapped Count"
        ylabel2 = "Lost Count"

    conf = data_dict["config"]
    #fig.suptitle(f"RF_ampltude = {conf["RF_amplitude"] dBm}")

    ax1.scatter(x, y1, marker="x")
    ax1.set_xlabel("Tickle Frequency (MHz)")
    ax1.set_ylabel(ylabel1)
    ax1.set_title("Trapped Signal")

    ax2.scatter(x, y2, marker="x")
    ax2.set_xlabel("Tickle Frequency (MHz)")
    ax2.set_ylabel(ylabel2)
    ax2.set_title("Lost Signal")

    plt.show()
```

```python
def preview(timestamp):
    data = load_data(timestamp)
    plot_data(data)
```

# Fitting

```python
def fit_gaussian(x, y):

    g = GaussianModel(prefix="g_") + Model(lambda x, c: c, prefix="b_")
    p = g.make_params()
    p.update(g.components[0].guess(y, x=x))

    # baseline: robust on edges
    p["b_c"].set(value=baseline_edges(y))

    # bounds: sigma>0, sigma<=scan_range, amp>=0
    dx = float(np.max(x) - np.min(x))
    p["g_sigma"].set(min=1e-9, max=dx)
    p["g_amplitude"].set(min=0)

    return g.fit(y, p, x=x)
```

```python
def fit_lorentzian(x, y):

    g = LorentzianModel(prefix="g_") + Model(lambda x, c: c, prefix="b_")
    p = g.make_params()
    p.update(g.components[0].guess(y, x=x))

    # baseline: robust on edges
    p["b_c"].set(value=baseline_edges(y))

    # bounds: sigma>0, sigma<=scan_range, amp>=0
    dx = float(np.max(x) - np.min(x))
    p["g_sigma"].set(min=1e-9, max=dx)
    p["g_amplitude"].set(min=0)

    return g.fit(y, p, x=x)
```

```python
def fit_voigt(x, y):

    g = VoigtModel(prefix="g_") + Model(lambda x, c: c, prefix="b_")
    p = g.make_params()
    p.update(g.components[0].guess(y, x=x))

    # baseline: robust on edges
    p["b_c"].set(value=baseline_edges(y))

    # bounds: widths>0, widths<=scan_range, amp>=0
    dx = float(np.max(x) - np.min(x))
    p["g_sigma"].set(min=1e-9, max=dx)
    p["g_gamma"].set(min=1e-9, max=dx)
    p["g_amplitude"].set(min=0)

    return g.fit(y, p, x=x)
```

```python
def fit_gaussian_peaks(data_dict):

    result = {}
    x = data_dict["arr_of_setpoints"]

    result["lost"] = fit_gaussian(x, data_dict["ratio_lost"])
    result["trapped"] = fit_gaussian(x, -data_dict["ratio_signal"])  # invert negative peak

    return result
```

```python
def fit_lorentzian_peaks(data_dict):

    result = {}
    x = data_dict["arr_of_setpoints"]

    result["lost"] = fit_lorentzian(x, data_dict["ratio_lost"])
    result["trapped"] = fit_lorentzian(x, -data_dict["ratio_signal"])  # invert negative peak

    return result
```

```python
def fit_voigt_peaks(data_dict):

    result = {}
    x = data_dict["arr_of_setpoints"]

    result["lost"] = fit_voigt(x, data_dict["ratio_lost"])
    result["trapped"] = fit_voigt(x, -data_dict["ratio_signal"])  # invert negative peak

    return result
```

```python
def fit_all(timestamp):

    results = {}
    data = load_data(timestamp)
    results["gaussian"] = fit_gaussian_peaks(data)
    results["lorentzian"] = fit_lorentzian_peaks(data)
    results["voigt"] = fit_voigt_peaks(data)

    return results
```
